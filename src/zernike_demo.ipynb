```python
s,w = dwt(np.random.normal(np.zeros((1024,1024)), 1.), 9)
plt.imshow(s)
```

```python
xgv=(np.arange(1024)+.5)/512-1.
ygv=(np.arange(1024)+.5)/512-1.
x,y=np.meshgrid(xgv,ygv)
rho=np.sqrt(x**2.+y**2.)
phi=np.atan2(y,x)
msk=(rho>=1.)
```

```python
clst,res=demo.zerndec(s[~msk],rho[~msk],phi[~msk],N=20)
```

```python
plt.semilogy(np.concatenate(clst)**2.)
plt.xlabel('J')
plt.ylabel('RMS')
```
