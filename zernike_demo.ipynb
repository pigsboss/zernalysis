```python
from pimms.filters import dwt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import zernalysis.base as base
from importlib import reload
reload(base)
```

```python
N=512
s,w=dwt(np.random.normal(np.zeros((N,N)), 1.), 7)
plt.imshow(s)
```

```python
xgv=(np.arange(N)+.5)*2/N-1.
ygv=(np.arange(N)+.5)*2/N-1.
x,y=np.meshgrid(xgv,ygv)
rho=np.sqrt(x**2.+y**2.)
phi=np.atan2(y,x)
msk=(rho>=1.)
```

```python
clst,res=base.zdec(s[~msk],rho[~msk],phi[~msk],N=20)
```

```python
plt.semilogy(clst**2.)
plt.xlabel('J')
plt.ylabel('RMS')
```
